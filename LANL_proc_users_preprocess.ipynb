## Data processing phase 1 for LANL proc.txt file 

In this notebook we do few things

1. Analysis of the redteam data.
2. extract events for each user who are also have events in the redteam.txt file

This should allow us to process smaller files further (see Data processing phase 2) for model input 


```python
import os

proc_fname = '/Volumes/TheFarSide/Data/LANL_data/proc.txt'
users_dir = '/Volumes/TheFarSide/Data/LANL_data/proc_users'
redteam_fname = '/Volumes/TheFarSide/Data/LANL_data/redteam.txt'
```

```python
# find out the number of users in the redteam.txt
users = {}

with open(redteam_fname, 'r') as red:
    for line in red:
        items = line.strip().split(',')
        user = items[1].strip().split('@')[0]
        if user in users:
            users[user] +=1
        else:
            users[user] = 1

print('num users in redteam.txt:', len(users.keys()))
#print(users)
        
```

```python
users['U748']
```

```python
if not os.path.exists(users_dir):
    os.makedirs(users_dir)


# create files for all users in redteam.txt
users_files = {}
for k,v in users.items():
    fname = '{0}/{1}.txt'.format(users_dir, k)
    f = open(fname, 'w')
    users_files[k] = f
```

```python
# process auth.txt for all users's events in the redteam.txt and extract their data
count = 0

with open(proc_fname, 'r') as proc:
    for line in proc:
        items = line.strip().split(',')
        user = items[1].strip().split('@')[0]
        if user in users_files:
            users_files[user].write(line)
        count+=1
        if (count % 1000000)==0:
            print('processed: {0} lines'.format(count))

```

```python
for _,v in users_files.items():
    v.close()
```
